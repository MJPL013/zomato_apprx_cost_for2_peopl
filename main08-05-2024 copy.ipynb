```python
df['cuisines'] = df['cuisines'].str.replace('Cafe', 'Cuisine Cafe')
df['rest_type'] = df['rest_type'].str.replace('Cafe', 'Rest Type Cafe')

df['cuisines'] = df['cuisines'].str.replace('Bakery', 'Cuisine Bakery')
df['rest_type'] = df['rest_type'].str.replace('Bakery', 'Rest Type Bakery')

df.head(5)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt 

k_values = list(range(1, 31)) # Adjust the range as needed

for k in k_values:
    knn_model = KNeighborsRegressor(n_neighbors=k)
    knn_model.fit(X_train, y_train)

    y_pred_val = knn_model.predict(X_val) 

    mse = mean_squared_error(y_val, y_pred_val)
    r2 = r2_score(y_val, y_pred_val)

    print(f'K = {k}, MSE: {mse:.4f}, R-squared: {r2:.4f}')
```

```python
k_values = list(range(1, 31)) # Adjust the range as needed

mse_values = []
r2_values = []

for k in k_values:
    knn_model = KNeighborsRegressor(n_neighbors=k)
    knn_model.fit(X_train, y_train)  # Train on training set

    y_pred_val = knn_model.predict(X_val)  # Predictions on validation set

    mse = mean_squared_error(y_val, y_pred_val)
    r2 = r2_score(y_val, y_pred_val)

    mse_values.append(mse)
    r2_values.append(r2)

# Visualization for selecting the best K
plt.figure(figsize=(10, 6))
plt.plot(k_values, mse_values, label='MSE')
# plt.plot(k_values, r2_values, label='R-squared')
plt.xlabel('K Value')
plt.ylabel('Performance Metric')
plt.title('KNN Performance on Validation Set')
plt.legend()
plt.show()

# Identify best K value
best_k = k_values[mse_values.index(min(mse_values))] 
print('Best K (considering MSE):', best_k)
```

```python
# Retrain model with best K
final_knn_model = KNeighborsRegressor(n_neighbors=best_k)
final_knn_model.fit(X_train, y_train)

# Predictions on validation set
y_pred_val = final_knn_model.predict(X_val)

# Calculate final metrics
mse = mean_squared_error(y_val, y_pred_val)
r2 = r2_score(y_val, y_pred_val)

print('KNN Performance on Validation Set (Best K):')
print('MSE:', mse)
print('R-squared:', r2)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score 
import math 

# Define Feature Sets
features_all = df.drop('approx_cost(for two people)', axis=1).columns  
features_no_votes = df.drop(['votes', 'approx_cost(for two people)'], axis=1).columns
features_no_review_count = df.drop(['Review_Count', 'approx_cost(for two people)'], axis=1).columns 
features_no_votes_no_reviews = df.drop(['votes', 'Review_Count', 'approx_cost(for two people)'], axis=1).columns

# KNN Model and Feature Comparison

r2_values = []
rmse_values = []

for case, features in enumerate([features_all, features_no_votes, features_no_review_count, features_no_votes_no_reviews]):
    print(f"----- Case {case+1} -----") 
    X_case = X[features]  
    print(features)

    k_values = list(range(1, 10)) 

    case_r2_values = []
    case_rmse_values = []

    for k in k_values:
        knn_model = KNeighborsRegressor(n_neighbors=k)
        knn_model.fit(X_train, y_train)  
        y_pred_val = knn_model.predict(X_val) 

        mse = mean_squared_error(y_val, y_pred_val)
        rmse = math.sqrt(mse)  # Calculate RMSE
        r2 = r2_score(y_val, y_pred_val)

        case_r2_values.append(r2)
        case_rmse_values.append(rmse)

        print(f'K = {k}, MSE: {mse:.2f}, RMSE: {rmse:.5f} R-squared: {r2:.4f}') 
        r2_values.append(case_r2_values)
        rmse_values.append(case_rmse_values)

    print()  

# Plot R2 values
for case, r2 in enumerate(r2_values):
    plt.plot(k_values, r2, label=f'Case {case+1}')
plt.xlabel('K')
plt.ylabel('R-squared')

plt.show()

# Plot RMSE values
for case, rmse in enumerate(rmse_values):
    plt.plot(k_values, rmse, label=f'Case {case+1}')
plt.xlabel('K')
plt.ylabel('RMSE')

plt.show()
```
